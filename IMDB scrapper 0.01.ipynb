```python
import requests as req
```

```python
page = req.get('https://www.imdb.com/title/tt0111161/')
```

```python
page
```

```python
page.content
```

```python
from bs4 import BeautifulSoup
```

```python
mysoup = BeautifulSoup(page.content,'html.parser')
```

```python
print(mysoup.prettify)
```

```python
myear = mysoup.find(id="titleYear").text
myear = myear.replace('(','')
myear = myear.replace(')','')
myear
```

```python
mtitle = mysoup.find(class_="originalTitle").text
```

```python
mrating = mysoup.find('span',itemprop='ratingValue').text
```

```python
mtime = mysoup.find('time').text
mtime = mtime.replace('\n','')
mtime = mtime.replace(" ",'')
```

```python
mgenre = mysoup.find(class_='subtext')
mgenre = mgenre.find('a').text
mgenre = mgenre.replace('\n','')
mgenre = mgenre.replace(' ','')
mgenre
```

```python
import pandas as pd
```

```python
mdict = {'Name':mtitle,'Year':myear,'Genre':mgenre,'Rating':mrating,'Time':mtime}
minfo = pd.DataFrame(mdict,index=[0])
```
