```python
import requests as req
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def pageextractor():
    for item in searchsoup.findAll(class_='lister-item-header'):
        try:
            item = 'https://www.imdb.com' + item.find('a').get('href')
            moviegetter(item)
        except:
            pass
        
def moviegetter(url):
    page = req.get(url)
    mysoup = BeautifulSoup(page.content,'html.parser')

    myear = mysoup.find(id="titleYear").find('a').text
    myear = myear.replace('(','')
    myear = myear.replace(')','')

    titletemp = mysoup.find(class_='title_wrapper').select('h1')[0].text.strip().split('\xa0')
    del titletemp[-1]

    mrating = mysoup.find('span',itemprop='ratingValue').text
    
    mtime = mysoup.find('time').text
    mtime = mtime.replace('\n','')
    mtime = mtime.replace(" ",'')
    
    mcountry = mysoup.find(id='titleDetails').findAll(class_='txt-block')[1].find('a').text
    #mcountry = mcountry.find('a').text

    mstory = mysoup.find(id='titleStoryLine').findAll('span')[1].text
    
    
    mdict = {'Name':titletemp[0],'Year':myear,'Country':mcountry,'Rating':mrating,'Time':mtime,'Story':mstory}
    dflist.append(mdict)
    
    #print(mdict)
    #print(type(mcountry))
    return dflist

```
