```python
def pageextractor():
    for item in searchsoup.findAll(class_='lister-item-header'):
        try:
            item = 'https://www.imdb.com' + item.find('a').get('href')
            moviegetter(item)
        except:
            pass
        
def moviegetter(url):
    page = req.get(url)
    mysoup = BeautifulSoup(page.content,'html.parser')

    myear = mysoup.find(id="titleYear").find('a').text
    myear = myear.replace('(','')
    myear = myear.replace(')','')

    titletemp = mysoup.find(class_='title_wrapper').select('h1')[0].text.strip().split('\xa0')
    del titletemp[-1]

    mrating = mysoup.find('span',itemprop='ratingValue').text
    mtime = mysoup.find('time').text
    mtime = mtime.replace('\n','')
    mtime = mtime.replace(" ",'')
    mgenre = mysoup.find(class_='subtext')
    mgenre = mgenre.find('a').text
    mgenre = mgenre.replace('\n','')
    mgenre = mgenre.replace(' ','')
    mgenre
    mdict = {'Name':titletemp[0],'Year':myear,'Genre':mgenre,'Rating':mrating,'Time':mtime}
    dflist.append(mdict)
    return dflist

```
